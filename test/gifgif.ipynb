```python
from IPython.display import Image, display
import os

# --- Configuration ---
# Assuming your GIFs are saved in the 'evaluation_gifs' directory
gif_directory = "evaluation_gifs"

# --- List all GIFs in the directory ---
gif_files = [f for f in os.listdir(gif_directory) if f.endswith('.gif')]

if not gif_files:
    print(f"No GIFs found in '{gif_directory}'. Please run your evaluation script first.")
else:
    print(f"Found {len(gif_files)} GIF(s) in '{gif_directory}':")
    for gif_file in gif_files:
        gif_path = os.path.join(gif_directory, gif_file)
        print(f"\nDisplaying: {gif_file}")
        try:
            # Display the GIF directly in the notebook output
            display(Image(filename=gif_path))
        except Exception as e:
            print(f"Could not display {gif_file}: {e}")
```
